```python
def current_page_URL(driver):
    medicine_posts = driver.find_elements(By.CLASS_NAME, "txtL")

    for post in medicine_posts:
        onclick_name = post.get_attribute('onclick')
        
        if onclick_name is not None:
            if onclick_name[:20] == 'drug_detailHref_mfds':
                url_parameter = onclick_name[22:-2]
                url = 'https://nedrug.mfds.go.kr/pbp/CCBBB01/getItemDetail?itemSeq=' + url_parameter
            elif onclick_name[:15] == 'drug_detailHref':
                url_parameter = onclick_name[17:-2]
                url = 'https://www.health.kr/searchDrug/result_drug.asp?drug_cd=' + url_parameter
            else:
                print(onclick_name)
                url = ''
            
            if url not in url_list and url != '':
                url_list.append(url)

            else:
                # 이미 들어가 있는 URL
                print(url)
```

```python
url_list = []
```

```python
for i in range(2, 6):

    # 약 게시물 선택
    current_page_URL(driver)
    
    # 페이지 넘김
    elem_xpath = f"/html/body/div/section/section/section/article[2]/div[2]/span/a[{i}]"
    map_icon = driver.find_element(By.XPATH, elem_xpath)
    map_icon.click()
    time.sleep(2)
```
